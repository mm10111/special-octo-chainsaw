```python
import transformers
import torch
```

```python
model = "tiiuae/falcon-7b"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline("text-generation", model=model, tokenizer=tokenizer, torch_dtype=torch.bfloat16, trust_remote_code=True, device_map="auto")
```

```python
sequences = pipeline("what do you know about machine learning", max_length=400, do_sample=True, top_k=10, num_return_sequences=1, eos_token_id=tokenizer.eos_token_id)
```

```python
for seq in sequences:
    print(f"RESULT: {seq['generated_text']}")
```
